```python
def encrypt_text(text, shift):
  ans = ""
  for char_val in text:
    if char_val == " ":
      ans += " "
    elif char_val.isupper():
      ans += chr(((ord(char_val) - ord('A') + shift) % 26) + ord('A'))
    else:
      ans += chr(((ord(char_val) - ord('a') - shift) % 26) + ord('a'))
  return ans


plain_text = "My name is Ranjeet"
shift_pattern = 3


print("Plain Text is:", plain_text)
print("Shift Pattern is:", shift_pattern)
print("Cipher Text is:", encrypt_text(plain_text, shift_pattern))
```
